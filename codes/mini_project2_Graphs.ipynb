```python
import numpy as np
import matplotlib.pyplot as plt

# creating the dataset
data = {
  'DBN': 0.07009095773140717,
  'PEO-DBN': 0.08295429932856833,
  'PBT-DBN': 0.208313194780087,
  'AFSA-GA-PSO-DBN': 0.06507615294493561,
}

models = list(data.keys())
values = list(data.values())

fig, ax = plt.subplots(figsize = (8, 4))

# creating the bar plot
plt.bar(models, values, color ='aquamarine', width = 0.2, hatch ='//', edgecolor="black")

# # Adjust the border color
# ax.hist(courses, values, edgecolor = "black")

plt.xlabel("ML Models")
plt.ylabel("FPR Metric")
plt.title("FPR Metric Simulation on Water Storage Tanks Dataset")
plt.show()
```

```python
# Define precision and recall
precision = 0.9389

# Calculate False Positive Rate (FPR)
FPR = (1 - precision) / precision

# Print the result
print("False Positive Rate (FPR):", FPR)
```

```python
import pandas as pd
from matplotlib import pyplot as plt

# creating the dataset
data = {
  'Logistic Regression': 0.6258,
  'KNN': 0.9903,
  'SVMLinear': 0.8649,
  'SVMRadial': 0.8680,
  'SVMPoly': 0.8765,
  'Decision Tree': 0.9867,
  'Random Forest': 0.9896,
  'XGBoost': 0.9923,
  'Bagging': 0.9887,
  'RNN': 0.8822,
  'LSTM': 0.8986,
  'BiLSTM': 0.8878,
  'TabNet': 0.9459,
  'DBN': 0.9467,
  'PEO-DBN': 0.8932,
  'PBT-DBN': 0.9276,
  'AFSA-GA-PSO-DBN': 0.9544,

}

models = list(data.keys())
values = list(data.values())

# Figure Size
fig, ax = plt.subplots(figsize =(16, 9))

# Horizontal Bar Plot
ax.barh(models, values, color='cornflowerblue', edgecolor="black")

# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
  ax.spines[s].set_visible(False)

# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')

# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)

# Add x, y gridlines
# ax.grid(b = True, color ='grey', linestyle ='-.', linewidth = 0.5, alpha = 0.2)

# Show top values
ax.invert_yaxis()

# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.005, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='black')

# Add Plot Title
# ax.set_title('Sports car and their price in crore', loc ='left', )

plt.xlim(0.6, 1.0)
# Show Plot
plt.show()
```
